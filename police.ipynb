```python
# Q. 1) Instruction ( For Data Cleaning ) - Remove the column that only contains missing values.
```

```python
data.head(3)
```

```python
# drop the column that has missing values
data.drop(labels=["country_name","search_type"], axis=1, inplace=True)
```

```python
data.head()
```

```python
# to clean all the dataset we have to know if there is still any Nan values
data.isnull().sum()
```

```python
# drop all null values
data.dropna(inplace=True)
```

```python
# Q. 4) Question ( mapping + data-type casting ) - What is the mean stop_duration ?
```

```python
data.head(2)
```

```python
data.stop_duration.value_counts()
```

```python
# to get the mean for the stop_duration
data['stop_duration'] = data['stop_duration'].map({"0-15 Min": 7.5, "16-30 Min": 23, "30+ Min": 37})
```

```python
data.head(15)
```

```python
# now we get the mean after o changed the values
data['stop_duration'].mean()
```
